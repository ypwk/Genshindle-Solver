```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df = pd.read_csv('./archive/genshin_updated.csv', encoding='ISO-8859-1')

# Step 3: Add a new column for version data
df['version'] = ''

response = requests.get('https://genshin-impact.fandom.com/wiki/Character/List')

soup = BeautifulSoup(response.text, 'html.parser')

characters_table = soup.find('table', class_='article-table sortable alternating-colors-table')
rows = characters_table.find_all('tr')

character_versions = {}
for row in rows[1:]:
    cols = row.find_all('td')
    if cols:
        a_tag = cols[0].find('a')
        character_name = a_tag['title'].strip()
        version = cols[-1].text.strip()
        character_versions[character_name] = version

for character_name, version in character_versions.items():
    df.loc[df['name'].apply(lambda x: x in character_name), 'version'] = version


df.to_csv('./archive/genshin_updated.csv', index=False, encoding='ISO-8859-1')
```

```python

def extract_last_word_talent(row):
    for col in ["talent_book_1-2","talent_book_2-3","talent_book_3-4","talent_book_4-5","talent_book_5-6","talent_book_6-7","talent_book_7-8","talent_book_8-9","talent_book_9-10"]:  # Add all your talent_book columns here
        if pd.notnull(row[col]) and row[col].strip() != '':
            return row[col].split()[-1]
    return None

df['Talent'] = df.apply(extract_last_word_talent, axis=1)
```
